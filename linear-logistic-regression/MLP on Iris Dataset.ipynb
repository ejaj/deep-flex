```python
import pandas as pd
import numpy as np
from keras.models import Sequential
from keras.layers import Dense, Activation
from keras.utils import to_categorical
```

```python
datatrain = pd.read_csv('data/iris_train.csv')
datatrain.head(5)
```

```python
datatrain.loc[datatrain['class'] == 'Iris-setosa', 'class'] = 0
datatrain.loc[datatrain['class'] == 'Iris-versicolor','class'] = 1
datatrain.loc[datatrain['class'] == 'Iris-virginica', 'class'] = 2
datatrain = datatrain.apply(pd.to_numeric)
```

```python
datatrain.head(5)
```

```python
datatrain_array = datatrain.values
datatrain_array
```

```python
x_train = datatrain_array[:,:4]
x_train
```

```python
y_train = datatrain_array[:,4]
y_train
```

```python
#change target format
y_train = to_categorical(y_train)
y_train
```

***Build a Sequential Keras Model***

Build a squential Keras model Multilayer perceptron model, with one hidden layer. Input layer : 4 neuron, represents the feature of Iris(Sepal Length etc) Hidden layer : 10 neuron, activation using ReLU Output layer : 3 neuron, represents the class of Iris, Softmax Layer

```python
model = Sequential()
model.add(Dense(units=10, input_dim=4))
model.add(Activation("relu"))
model.add(Dense(units=3))
model.add(Activation("softmax"))
```

```python
#Compile the model and choose an optimizer and loss function for training and optimizing your data.
#optimizer = stochastic gradient descent with no batch-size
#loss function = categorical cross entropy
#learning rate = default from keras.optimizer.SGD, 0.01
model.compile(loss='categorical_crossentropy', optimizer='sgd', metrics=['accuracy'])
```

```python
#Train the model using the datatrain‘model.fit’ function.
model.fit(x_train, y_train, epochs=100, batch_size=120)
```
